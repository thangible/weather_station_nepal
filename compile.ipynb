```python
import os
import numpy as np
import pandas as pd
pd.set_option('display.max_rows', None)
```

```python
dir = os.getcwd()
```

```python
temp_spec = [(0,6),(6,12),(12,18)]
temp_cols = ["day_of_year","tmax","tmin"]
temp_dir = os.path.join(dir,"weatherstationNepal","Temperature")
pre_spec = [(0,7),(7,14)]
pre_cols = ["day_of_year","pre"]
pre_dir = os.path.join(dir,"weatherstationNepal","Precipitation")
output_cols = ["year","month","day","v4","tmax","tmin","pre","v8","v9"]
output_format = '%6s%4s%4s%6s%6s%6s%6s%6s%6s'
output_directory = os.path.join(dir,"output")
```

```python
def compile(station_id, year):
    #FORMAT STATION ID
    station_id = str(station_id).zfill(4)
    
    #SET PATH AND NAME
    OUTPUT_NAME = str(station_id) + str(year)[0:2] + "." + str(year)[2:]
    OUTPUT_SUBDIR = os.path.join(output_directory,station_id)
    AS_NAME = "AS" + str(station_id) + str(year)[0:2] + "." + str(year)[2:]
    TEMP_NAME = "TA" + str(station_id) + str(year)[0:2] + "." + str(year)[2:]          
    AS_PATH = os.path.join(dir,"weatherstationNepal","Precipitation",str(station_id),AS_NAME )
    TEMP_PATH = os.path.join(dir,"weatherstationNepal","Temperature",str(station_id),TEMP_NAME)
    
    if not(os.path.isfile(AS_PATH)):
        print(f"precipitation path does not exists for station {station_id} and year {year} \npath: {AS_PATH}")
    elif not(os.path.isfile(TEMP_PATH)):
        print(f"temperature path does not exists for station {station_id} and year {year} \npath: {TEMP_PATH}")
    else:
        #READ
        AS_FILE = pd.read_fwf(AS_PATH,skiprows=0,skipfooter=1,colspecs=pre_spec,names=pre_cols)
        TEMP_FILE = pd.read_fwf(TEMP_PATH,skiprows=2, skipfooter=1,colspecs=temp_spec,names=temp_cols)

        #MERGE
        output = pd.merge(AS_FILE,TEMP_FILE,on="day_of_year")

        #DEFINE DATE
        output["year"] = year
        output["date"] = pd.to_datetime(output["year"]*1000 + output["day_of_year"], format = "%Y%j")
        output["day"] = output["date"].dt.day
        output["month"] = output["date"].dt.month

        #ADD DUMMY VARS
        output['v4'] = 0
        output['v8'] = 0
        output['v9'] = 0

        #DROP COLS
        output.drop(columns=["day_of_year","date"])

        #REARRAGE COLS
        output = output.reindex(columns=output_cols)
        
        #CREATE OUTPUT DIR IF NOT THERE
        if not os.path.exists(OUTPUT_SUBDIR):
            os.makedirs(OUTPUT_SUBDIR)
        
        #SAVE
        np.savetxt(os.path.join(OUTPUT_SUBDIR,OUTPUT_NAME),output.values, fmt = output_format)

        return output
        
```

```python
compile(104,1984)
```

```python
temp_stations = [name for name in os.listdir(temp_dir) if os.path.isdir(os.path.join(temp_dir,name))]
for station_id in temp_stations:
    for year in range(1950,2015):
        compile(station_id,year)
```

```python
pre_stations = [name for name in os.listdir(pre_dir) if os.path.isdir(os.path.join(pre_dir,name))]
for station_id in pre_stations:
    for year in range(1950,2015):
        compile(station_id,year)
```
